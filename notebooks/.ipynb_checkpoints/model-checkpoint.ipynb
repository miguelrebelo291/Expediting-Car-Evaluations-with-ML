## Column Transformers

```python
class RareLabelGrouper(BaseEstimator, TransformerMixin):
    def __init__(self, min_prop=0.005, column=None):
        self.min_prop = min_prop
        self.column = column
        self.keep_values_ = None
    def fit(self, X, y=None):
        s = X if isinstance(X, pd.Series) else pd.Series(X.iloc[:,0] if hasattr(X, "iloc") else X[:,0])
        vc = s.value_counts(dropna=False)
        total = len(s)
        keep = vc[vc / total >= self.min_prop].index
        self.keep_values_ = set(keep.tolist())
        return self
    def transform(self, X):
        s = X if isinstance(X, pd.Series) else pd.Series(X.iloc[:,0] if hasattr(X, "iloc") else X[:,0])
        return pd.DataFrame(np.where(s.isin(self.keep_values_), s, "Other"),
                            columns=[self.column if self.column else "col"])

class ColumnSelector(BaseEstimator, TransformerMixin):
    def __init__(self, column): self.column = column
    def fit(self, X, y=None): return self
    def transform(self, X): return pd.DataFrame(X[self.column])

class CustomWinsorizer(BaseEstimator, TransformerMixin):
    def __init__(self, variables, lower_bound=0.01, upper_bound=0.99):
        self.variables = variables
        self.lower_bound = lower_bound
        self.upper_bound = upper_bound
        self.limits_ = {}

    def fit(self, X, y=None):
        for var in self.variables:
            if var in X.columns:
                lower_limit = X[var].quantile(self.lower_bound)
                upper_limit = X[var].quantile(self.upper_bound)
                self.limits_[var] = (lower_limit, upper_limit)
        return self

    def transform(self, X):
        X_transformed = X.copy()
        for var, (lower_limit, upper_limit) in self.limits_.items():
            if var in X.columns:
                X_transformed[var] = np.clip(X_transformed[var], lower_limit, upper_limit)
        return X_transformed
```

```python
# BaseEstimator → dá compatibilidade com pipelines, grid-search e clonagem
# TransformerMixin → fornece fit_transform() automaticamente
class Mode_Imput_By_Brand(BaseEstimator, TransformerMixin):

    """
    O objetivo com este transformer é preencher os valores nulos de cada coluna através da moda, mas primeiro, agrupar por brand.
    Como fallback, caso nao haja nenhuma linha preenchida daquela brand em especifico, depois de utilizar este transformer
    iremos utilzar o SimpleImputer.
    
    """
    
    def __init__(self, imput_cols = None , group_col = "brand"):       
        self.imput_cols = imput_cols
        self.group_col = group_col

    # é obrigatorio colocar X e y em todos os metodos de fit do sklearn
    def fit(self , X, y = None):
        X = X.copy()  # em qualquer metodo do sklearn criamos uma copia do df original, pois no pipeline é usado o mesmo df nao modificado para os vários passos
        
        if self.imput_cols is None: # sklearn nao gosta de args obrigatórios entao fazemos isto para contornar
            self.imput_cols = [col for col in X.columns if col != self.group_col]
        
        self.mode_maps_ = {}

        for col in self.imput_cols:
            mode_map = X.groupby(self.group_col)[col].agg(lambda x: x.mode().iloc[0]) 
            self.mode_maps_[col] = mode_map 

        return self # o .fit do sklearn retorna sempre o self
    
    def transform(self, X):
        X = X.copy()

        for col in self.imput_cols:
            X[col] = X[col].fillna(X[self.group_col].map(self.mode_maps_[col]))

        return X
```

## Scalling Data

Scaling

- **Why scale:** Linear/regularized models are scale-sensitive. Numeric features can have very different ranges (ex: mileage vs engine size), which biases coefficients.

- **Choice:** **StandardScaler** for numeric columns (after imputation and outlier capping).

- **StandardScaler over MinMaxScaler:** Standardization centers and normalizes variance—well-aligned with Ridge/Lasso penalties and keeps One-Hot Enconding interpretable. All scaling is fit on the training fold only within the pipeline to ensure that there are **no leakage**.

```python
scaler = StandardScaler()
```

## Feature Selection with Cross-Validation

### Approach 1: RFECV (Recursive Feature Elimination with Cross-Validation)

Uses RFECV to automatically select the optimal number of features through internal cross-validation.

```python
# MAE scorer (sklearn maximizes, so we negate MAE to minimize)
mae_scorer = make_scorer(mean_absolute_error, greater_is_better=False)

# Cross-validation: split data into 5 parts
cv = KFold(n_splits=5, shuffle=True, random_state=44)

# RFECV recursively eliminates features and chooses the best number via CV
rfecv = RFECV(
    estimator=LinearRegression(),
    step=1,                    # Remove 1 feature at a time
    cv=cv,
    scoring=mae_scorer,
    min_features_to_select=5,  # Minimum features to keep
    n_jobs=-1
)

# Pipeline: preprocessing → feature selection
pipe = Pipeline([
    ("prep", preprocessor),
    ("rfecv", rfecv)
])

# Training
print("Training RFECV...") 
pipe.fit(X_train, y_train_transformed)

# Extract results
optimal_n = pipe.named_steps['rfecv'].n_features_
cv_scores = -pipe.named_steps['rfecv'].cv_results_['mean_test_score']

# The array has different size than number of features
best_mae = cv_scores.min()  # Best MAE in array
best_idx = cv_scores.argmin()  # Index of best MAE

print(f"\nResults:")
print(f"   Optimal features: {optimal_n}")
print(f"   MAE (CV): {best_mae:.4f}")
print(f"   Total configurations tested: {len(cv_scores)}")
```

```python
# Visualization
plt.figure(figsize=(10, 6))

# Calculate number of features for each point
# RFECV starts with max_features and removes step at each iteration
max_features = len(cv_scores) + rfecv.min_features_to_select - 1
n_features_array = range(rfecv.min_features_to_select, max_features + 1)

plt.plot(n_features_array, cv_scores, 'o-', linewidth=2, markersize=4)
plt.axvline(optimal_n, color='red', linestyle='--', 
            label=f'Optimal: {optimal_n} features (MAE: {best_mae:.4f})')
plt.xlabel('Number of Features')
plt.ylabel('MAE (Cross-Validation)')
plt.title('RFECV: Automatic Feature Selection')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

### Approach 2: Manual RFE with Cross-Validation

Tests different numbers of features using cross-validation for each configuration.

```python
# MAE scorer
mae_scorer = make_scorer(mean_absolute_error, greater_is_better=False)

# Cross-validation strategy
cv_strategy = KFold(n_splits=5, shuffle=True, random_state=44)

# Strategy: more granular near the expected optimum
n_features_to_test = (
    list(range(10, 50, 10)) +      # [10, 20, 30, 40]
    list(range(50, 68, 2))          # [50, 52, 54, ..., 66]
)

print(f"Testing {len(n_features_to_test)} feature configurations")
print(f"Range: {min(n_features_to_test)} to {max(n_features_to_test)} features\n")

# Testing different numbers of features
cv_results = []

for i, n_feat in enumerate(n_features_to_test, 1):
    print(f"[{i}/{len(n_features_to_test)}] Testing {n_feat} features...", end=' ')
    
    # Pipeline: preprocessing → RFE → model
    pipe = Pipeline([
        ("prep", preprocessor),
        ("rfe", RFE(LinearRegression(), n_features_to_select=n_feat)),
        ("model", LinearRegression())  # ✅ Now correct (RFE only selects)
    ])

        # Cross-validation on training set
    cv_scores = cross_val_score(
        pipe, X_train, y_train_transformed,
        cv=cv_strategy,
        scoring=mae_scorer,
        n_jobs=-1
    )
    cv_mae = -cv_scores  # Convert to positive MAE
    
    # Also evaluate on validation set
    pipe.fit(X_train, y_train_transformed)
    val_pred_log = pipe.predict(X_val)
    val_pred_original = np.expm1(val_pred_log)
    val_mae = mean_absolute_error(y_val, val_pred_original)    
    
    # Store results
    cv_results.append({
        'n_features': n_feat,
        'cv_mae_mean': cv_mae.mean(),
        'cv_mae_std': cv_mae.std(),
        'val_mae': val_mae
    })
    
    print(f"CV: {cv_mae.mean():.4f} | Val: {val_mae:.0f}")

# Results
cv_results_df = pd.DataFrame(cv_results)

# Find best configuration
best_idx = cv_results_df['cv_mae_mean'].idxmin()
best_n = cv_results_df.loc[best_idx, 'n_features']
best_cv_mae = cv_results_df.loc[best_idx, 'cv_mae_mean']
best_val_mae = cv_results_df.loc[best_idx, 'val_mae']

print("\n" + "="*80)
print("RESULTS")
print("="*80)
print(cv_results_df.to_string(index=False))

print("\n" + "="*80)
print("BEST CONFIGURATION")
print("="*80)
print(f"   Features: {int(best_n)}")
print(f"   MAE (CV): {best_cv_mae:.4f}")
print(f"   MAE (Val): {best_val_mae:.0f}")
print("="*80)
```

```python
# Visualization
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: CV MAE with error bars
ax1 = axes[0]
ax1.errorbar(cv_results_df['n_features'], 
             cv_results_df['cv_mae_mean'],
             yerr=cv_results_df['cv_mae_std'],
             marker='o', capsize=5, linewidth=2, markersize=6)
ax1.axvline(best_n, color='red', linestyle='--', 
            label=f'Optimal: {int(best_n)} features')
ax1.set_xlabel('Number of Features', fontsize=11)
ax1.set_ylabel('MAE (Cross-Validation)', fontsize=11)
ax1.set_title('CV: MAE vs Number of Features', fontsize=12, fontweight='bold')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot 2: CV vs Validation comparison
ax2 = axes[1]
ax2.plot(cv_results_df['n_features'], cv_results_df['cv_mae_mean'], 
         'o-', linewidth=2, markersize=6, label='CV MAE')
ax2_twin = ax2.twinx()  # Second Y axis for Validation MAE
ax2_twin.plot(cv_results_df['n_features'], cv_results_df['val_mae'], 
              's-', linewidth=2, markersize=6, color='coral', label='Validation MAE')
ax2.axvline(best_n, color='red', linestyle='--', alpha=0.5)
ax2.set_xlabel('Number of Features', fontsize=11)
ax2.set_ylabel('MAE (CV)', fontsize=11, color='steelblue')
ax2_twin.set_ylabel('MAE (Validation)', fontsize=11, color='coral')
ax2.set_title('Comparison: CV vs Validation', fontsize=12, fontweight='bold')
ax2.tick_params(axis='y', labelcolor='steelblue')
ax2_twin.tick_params(axis='y', labelcolor='coral')
ax2.grid(True, alpha=0.3) 

# Combine legends
lines1, labels1 = ax2.get_legend_handles_labels()
lines2, labels2 = ax2_twin.get_legend_handles_labels()
ax2.legend(lines1 + lines2, labels1 + labels2, loc='best')

plt.tight_layout()
plt.show()
```

### Train Final Model with Best Number of Features

```python
# Train final model with optimal number of features
final_model = Pipeline([
    ("prep", preprocessor),
    ("rfe", RFE(estimator=LinearRegression(), n_features_to_select=int(best_n))),
    ("model", LinearRegression()),
])

final_model.fit(X_train, y_train_transformed)

# Predictions
val_pred_log = final_model.predict(X_val)
val_pred = np.expm1(val_pred_log)

# Metrics
mae = mean_absolute_error(y_val, val_pred)
rmse = np.sqrt(mean_squared_error(y_val, val_pred))
r2 = r2_score(y_val, val_pred)

print("\n" + "="*80)
print(f"FINAL MODEL - {int(best_n)} FEATURES")
print("="*80)
print(f"\nValidation Set Performance:")
print(f"   MAE:  {mae:>10,.2f}")
print(f"   RMSE: {rmse:>10,.2f}")
print(f"   R²:   {r2:>10.4f}")
print("="*80)
```

# 3. Modeling and Evaluation with Cross-Validation

```python
print("\n" + "="*80)
print("MODEL COMPARISON")
print("="*80)

models = {
    "Linear Regression": LinearRegression(),
    "Ridge": RidgeCV(alphas=np.logspace(-3, 3, 50), cv=5),
    "Lasso": LassoCV(alphas=np.logspace(-3, 3, 50), cv=5, 
                     random_state=44, max_iter=5000),
    "ElasticNet": ElasticNetCV(alphas=np.logspace(-3, 3, 50), cv=5, 
                               random_state=44, max_iter=5000),
    "Decision Tree": DecisionTreeRegressor(random_state=44, max_depth=10)
}

results = []
fitted_models = {}

print("\nEvaluating models...\n")

for name, model in models.items():
    print(f"   • {name}...", end=' ')
        
    # Pipeline
    pipe = Pipeline([("prep", preprocessor), ("model", model)])    

        
    # Cross-validation
    cv_scores = cross_validate(
        pipe, X_train, y_train_transformed,
        cv=cv_strategy,
        scoring={'mae': mae_scorer, 'r2': 'r2'},
        return_train_score=True,
        n_jobs=-1
    )
    
    cv_mae = -cv_scores['test_mae']
    cv_mae_train = -cv_scores['train_mae']
    
    # Validation set
    pipe.fit(X_train, y_train_transformed)
    val_pred = np.expm1(pipe.predict(X_val))
    train_pred = np.expm1(pipe.predict(X_train))
    
    val_mae = mean_absolute_error(y_val, val_pred)
    train_mae = mean_absolute_error(y_train, train_pred)
    val_r2 = r2_score(y_val, val_pred)
        
    # Store results
    results.append({
        'Model': name,
        'CV_MAE': cv_mae.mean(),
        'CV_Std': cv_mae.std(),
        'Train_MAE': train_mae,
        'Val_MAE': val_mae,
        'Val_R2': val_r2,
        'Overfit': val_mae - train_mae
    })
       
    fitted_models[name] = pipe
    print(f"✓ (CV MAE: {cv_mae.mean():.4f})")

# Results DataFrame
results_df = pd.DataFrame(results).sort_values('CV_MAE')

print("\n" + "="*80)
print("RESULTS")
print("="*80)
print(results_df.to_string(index=False, float_format=lambda x: f'{x:.4f}'))

# Best model
best = results_df.iloc[0]

print("\n" + "="*80)
print(f"BEST MODEL: {best['Model']}")
print("="*80)
print(f"   CV MAE:  {best['CV_MAE']:.4f} (±{best['CV_Std']:.4f})")
print(f"   Val MAE: {best['Val_MAE']:.2f}")
print(f"   Val R²:  {best['Val_R2']:.4f}")
print("="*80)

# Visualizations
# Setup for plots
best_model = fitted_models[best['Model']]
best_pred = np.expm1(best_model.predict(X_val))
residuals = y_val - best_pred


# 3 main plots
fig = plt.figure(figsize=(16, 5))

# Plot 1: Model Comparison (CV MAE)
ax1 = plt.subplot(131)
colors = ['green' if i == 0 else 'steelblue' for i in range(len(results_df))]
ax1.barh(range(len(results_df)), results_df['CV_MAE'], 
         xerr=results_df['CV_Std'], capsize=5, color=colors, alpha=0.7)
ax1.set_yticks(range(len(results_df)))
ax1.set_yticklabels(results_df['Model'])
ax1.set_xlabel('MAE (Cross-Validation)', fontsize=11)
ax1.set_title('Model Comparison', fontsize=12, fontweight='bold')
ax1.grid(axis='x', alpha=0.3)
ax1.invert_yaxis()

# Plot 2: Actual vs Predicted (Best Model)
ax2 = plt.subplot(132)
ax2.scatter(y_val, best_pred, alpha=0.5, s=20)
lims = [min(y_val.min(), best_pred.min()), max(y_val.max(), best_pred.max())]
ax2.plot(lims, lims, 'r--', lw=2, label='Perfect')
ax2.set_xlabel('Actual Price', fontsize=11)
ax2.set_ylabel('Predicted Price', fontsize=11)
ax2.set_title(f'Predictions - {best["Model"]}\nR² = {best["Val_R2"]:.4f}', 
              fontsize=12, fontweight='bold')
ax2.legend()
ax2.grid(True, alpha=0.3)

# Plot 3: Residuals
ax3 = plt.subplot(133)
ax3.scatter(best_pred, residuals, alpha=0.5, s=20)
ax3.axhline(y=0, color='r', linestyle='--', lw=2)
ax3.set_xlabel('Predicted Price', fontsize=11)
ax3.set_ylabel('Residuals', fontsize=11)
ax3.set_title('Residual Analysis', fontsize=12, fontweight='bold')
ax3.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
